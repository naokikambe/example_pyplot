### Example 1: Simple Bar Chart

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 24, 20, 32]

plt.bar(categories, values, color='blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Chart')
plt.show()
```

### Example 2: Line Chart

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 15]

plt.plot(x, y, marker='o', linestyle='-', color='green')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Chart')
plt.show()
```

### Example 3: Combined Bar and Line Chart

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 24, 20, 32]
line_data = [10, 15, 8, 25]

fig, ax1 = plt.subplots()

ax1.bar(categories, values, color='blue', label='Bar Chart')
ax1.set_xlabel('Categories')
ax1.set_ylabel('Values', color='blue')
ax1.tick_params('y', colors='blue')

ax2 = ax1.twinx()
ax2.plot(categories, line_data, color='green', marker='o', linestyle='-', label='Line Chart')
ax2.set_ylabel('Line Data', color='green')
ax2.tick_params('y', colors='green')

plt.title('Combined Bar and Line Chart')
plt.legend()
plt.show()
```

### Example 4: Stacked Bar Chart

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values1 = [15, 24, 20, 32]
values2 = [8, 12, 15, 10]

plt.bar(categories, values1, label='Value 1')
plt.bar(categories, values2, bottom=values1, label='Value 2')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')
plt.legend()
plt.show()
```

### Example 5: Grouped Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values1 = [15, 24, 20, 32]
values2 = [8, 12, 15, 10]

bar_width = 0.35
index = np.arange(len(categories))

plt.bar(index, values1, width=bar_width, label='Value 1')
plt.bar(index + bar_width, values2, width=bar_width, label='Value 2')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Chart')
plt.xticks(index + bar_width / 2, categories)
plt.legend()
plt.show()
```

### Example 6: Combined Bar, Line, and Stacked Bar Chart

```python
import matplotlib.pyplot as plt

years = ['2023', '2024', '2045', '2076']
bar_values = [15, 24, 20, 32]
line_values = [10, 15, 8, 25]
stacked_values1 = [8, 12, 15, 10]
stacked_values2 = [5, 10, 12, 8]

fig, ax1 = plt.subplots()

# Bar and Line on the same axis
ax1.bar(years, bar_values, color='blue', label='Bar Chart')
ax1.set_xlabel('Years')
ax1.set_ylabel('Bar Values', color='blue')
ax1.tick_params('y', colors='blue')

ax2 = ax1.twinx()
ax2.plot(years, line_values, color='green', marker='o', linestyle='-', label='Line Chart')
ax2.set_ylabel('Line Values', color='green')
ax2.tick_params('y', colors='green')

# Stacked Bar on the same axis
ax1.bar(years, stacked_values1, bottom=bar_values, label='Stacked Value 1', color='orange')
ax1.bar(years, stacked_values2, bottom=[i+j for i,j in zip(bar_values, stacked_values1)], label='Stacked Value 2', color='red')

plt.title('Combined Bar, Line, and Stacked Bar Chart')

ax1.legend(loc='upper left', bbox_to_anchor=(0.1, 1))
ax2.legend(loc='upper right', bbox_to_anchor=(0.9, 1))
plt.show()
```

### Example 7: Combined Stacked Bar Chart Line Chart with External CSV Data

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data from the CSV file using pandas
df = pd.read_csv('sample_data.csv')

# Display the loaded data
print(df.head())

# Extract data for plotting
years = [ str(n) for n in df['Year'] ]
bar_values = df['Bar_Value']
line_values = df['Line_Value']
stacked_values1 = df['Stacked_Value1']
stacked_values2 = df['Stacked_Value2']

fig, ax1 = plt.subplots()

# Bar and Line on the same axis
ax1.bar(years, bar_values, color='blue', label='Bar Chart')
ax1.set_xlabel('Years')
ax1.set_ylabel('Bar Values', color='blue')
ax1.tick_params('y', colors='blue')

ax2 = ax1.twinx()
ax2.plot(years, line_values, color='green', marker='o', linestyle='-', label='Line Chart')
ax2.set_ylabel('Line Values', color='green')
ax2.tick_params('y', colors='green')

# Stacked Bar on the same axis
ax1.bar(years, stacked_values1, bottom=bar_values, label='Stacked Value 1', color='orange')
ax1.bar(years, stacked_values2, bottom=[i+j for i,j in zip(bar_values, stacked_values1)], label='Stacked Value 2', color='red')

plt.title('Combined Bar, Line, and Stacked Bar Chart with External CSV Data')

ax1.legend(loc='upper left', bbox_to_anchor=(0.1, 1))
ax2.legend(loc='upper right', bbox_to_anchor=(0.9, 1))
plt.show()
```

### Sine curve

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 2*pi
x = np.linspace(0, 2 * np.pi, 1000)

# Calculate y values for the sine function
y = np.sin(x)

# Plot the sine curve
plt.plot(x, y, label='sin(x)')
plt.title('Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.text(1, 0.8, r'$y = \sin(x)$', fontsize=12, color='blue')

plt.show()
```

### Quadratic Curve

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from -2 to 2
x = np.linspace(-2, 2, 1000)

# Calculate y values for the quadratic function
y_quadratic = x**2 + x + 1

# Plot the quadratic curve
plt.figure(figsize=(8, 5))  # Adjust figure size if needed
plt.plot(x, y_quadratic, label='x^2 + x + 1', color='red')
plt.title('Quadratic Curve')
plt.xlabel('x')
plt.ylabel('x^2 + x + 1')
plt.legend()
plt.grid(True)
plt.text(1, 6, r'$y = x^2 + x + 1$', fontsize=12, color='red')

plt.show()
```
